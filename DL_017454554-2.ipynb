```python
import torch
from torch.utils.data import DataLoader, Dataset
from transformers import BertTokenizer, BertForQuestionAnswering, AdamW
from datasets import load_dataset
from tqdm import tqdm
from evaluate import load as load_metric
import random
from IPython.core.display import display, HTML

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# ===============================
# Step 1: Load SQuAD 2.0 Dataset
# ===============================
squad_dataset = load_dataset("squad_v2")

# Use a subset for training
sample_size = 1000  # 1000 same size to tain my model.
train_data = squad_dataset["train"].shuffle(seed=42).select(range(sample_size))
```

```python
# ===============================
# Step 2: Display some data
# ===============================

for i in range(5):  # Display 5 samples
    sample = train_data[i]
    print(f"Sample {i+1}:")
    print(f"Question: {sample['question']}")
    print(f"Context: {sample['context'][:200]}...")  # Truncated for readability
    print(f"Answer: {sample['answers']['text']}")
    print("-" * 80)
```

```python
# ===============================
# Step 3: Preprocess the Dataset
# ===============================

class QnADataset(Dataset):
    def __init__(self, data, tokenizer, max_len=512):
        self.data = data
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        sample = self.data[idx]
        context = sample["context"]
        question = sample["question"]
        answers = sample["answers"]["text"]

        # Tokenize context and question
        encoding = self.tokenizer(
            question,
            context,
            max_length=self.max_len,
            truncation=True,
            padding="max_length",
            return_offsets_mapping=True,
            return_tensors="pt",
        )

        offsets = encoding["offset_mapping"][0]
        input_ids = encoding["input_ids"].squeeze(0)
        attention_mask = encoding["attention_mask"].squeeze(0)

        # Find the answer span in tokenized format
        if answers:
            answer_text = answers[0]
            start_char = context.find(answer_text)
            end_char = start_char + len(answer_text)

            # Map character positions to token positions
            start_idx = end_idx = 0
            for idx, (start, end) in enumerate(offsets):
                if start <= start_char < end:
                    start_idx = idx
                if start < end_char <= end:
                    end_idx = idx
        else:
            start_idx = 0
            end_idx = 0

        return {
            "input_ids": input_ids,
            "attention_mask": attention_mask,
            "start_positions": torch.tensor(start_idx, dtype=torch.long),
            "end_positions": torch.tensor(end_idx, dtype=torch.long),
        }

from transformers import BertTokenizerFast, BertForQuestionAnswering, AdamW
# Load tokenizer and create dataset
tokenizer = BertTokenizerFast.from_pretrained("bert-base-uncased")
qna_dataset = QnADataset(train_data, tokenizer)
train_loader = DataLoader(qna_dataset, batch_size=8, shuffle=True)

# ===============================
# Step 4: Initialize BERT Model
# ===============================

model = BertForQuestionAnswering.from_pretrained("bert-base-uncased")
model.to(device)

optimizer = AdamW(model.parameters(), lr=3e-5)

# ===============================
# Step 5: Train the Model
# ===============================

epochs = 5
for epoch in range(epochs):
    model.train()
    loop = tqdm(train_loader, leave=True)
    for batch in loop:
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        start_positions = batch["start_positions"].to(device)
        end_positions = batch["end_positions"].to(device)

        optimizer.zero_grad()
        outputs = model(
            input_ids=input_ids,
            attention_mask=attention_mask,
            start_positions=start_positions,
            end_positions=end_positions,
        )
        loss = outputs.loss
        loss.backward()
        optimizer.step()

        loop.set_description(f"Epoch {epoch + 1}")
        loop.set_postfix(loss=loss.item())

# ===============================
# Step 6: Inference and Post-Processing
# ===============================

def clean_prediction(input_ids, tokenizer, start_idx, end_idx):
    if start_idx > end_idx or start_idx < 0 or end_idx >= len(input_ids):
        return ""
    return tokenizer.decode(input_ids[start_idx:end_idx + 1], skip_special_tokens=True)

# Test the model with a single sample
model.eval()
sample = train_data[0]
inputs = tokenizer(
    sample["question"],
    sample["context"],
    return_tensors="pt",
    max_length=512,
    truncation=True,
    padding="max_length",
).to(device)

outputs = model(**inputs)
start_idx = torch.argmax(outputs.start_logits)
end_idx = torch.argmax(outputs.end_logits)
pred_answer = clean_prediction(inputs["input_ids"][0], tokenizer, start_idx, end_idx)

print("\n--- Inference Results ---")
print(f"Question: {sample['question']}")
print(f"Ground Truth: {sample['answers']['text'][0] if sample['answers']['text'] else 'N/A'}")
print(f"Predicted Answer: {pred_answer}")
```

```python
predictions = []
references = []

for sample in train_data:
    if not sample["answers"]["text"]:  # Skip samples with no ground truth answer
        continue

    inputs = tokenizer(
        sample["question"],
        sample["context"],
        return_tensors="pt",
        max_length=512,
        truncation=True,
        padding="max_length",
    ).to(device)

    outputs = model(**inputs)
    start_idx = torch.argmax(outputs.start_logits)
    end_idx = torch.argmax(outputs.end_logits)
    pred_answer = clean_prediction(inputs["input_ids"][0], tokenizer, start_idx, end_idx)

    predictions.append({"id": sample["id"], "prediction_text": pred_answer})
    references.append({"id": sample["id"], "answers": sample["answers"]})

squad_metric = load("squad")
# Evaluate
results = squad_metric.compute(predictions=predictions, references=references)
print("\n--- Evaluation Results ---")
print(f"Exact Match: {results['exact_match']:.2f}")
print(f"F1 Score: {results['f1']:.2f}")
```

```python
def evaluate_multiple_samples(model, tokenizer, data, num_samples=5):
    model.eval()
    results = []

    for i, sample in enumerate(data):  # Enumerate the dataset directly
        if i >= num_samples:  # Stop after processing `num_samples`
            break

        inputs = tokenizer(
            sample["question"],
            sample["context"],
            return_tensors="pt",
            max_length=512,
            truncation=True,
            padding="max_length",
        ).to(device)

        outputs = model(**inputs)
        start_idx = torch.argmax(outputs.start_logits)
        end_idx = torch.argmax(outputs.end_logits)

        pred_answer = clean_prediction(inputs["input_ids"][0], tokenizer, start_idx, end_idx)
        ground_truth = sample["answers"]["text"][0] if sample["answers"]["text"] else "N/A"

        # Store results
        results.append({
            "question": sample["question"],
            "context": sample["context"][:200] + "...",  # Show part of the context
            "predicted_answer": pred_answer,
            "ground_truth": ground_truth,
        })

    # Display results
    for idx, res in enumerate(results):
        print(f"\nSample {idx + 1}")
        print(f"Question: {res['question']}")
        print(f"Context: {res['context']}")
        print(f"Predicted Answer: {res['predicted_answer']}")
        print(f"Ground Truth: {res['ground_truth']}")

# Call the function to evaluate 15 random samples
evaluate_multiple_samples(model, tokenizer, train_data, num_samples=15)
```
